## Απορία για την args της quad

```python
from scipy.integrate import quad
from scipy.special import jn

def integrand(x, n):
    """
    Bessel function of first kind and order n. 
    """
    return jn(n, x)


x_lower = 0  # το κατώτερο όριο του x
x_upper = 10 # το ανώτατο όριο του x

val, abserr = quad(integrand, x_lower, x_upper, args=(3,))

print(val, abserr)
```

Η συνάρτηση που δίνουμε στη quad πρέπει να έχει πρώτο όρισμα τη μεταβλητή ολοκλήρωσης.
Εδώ η τιμή args=(3,).
Αντιστοιχεί στην δευτερη μεταβλητή. 
Αν την ορίζαμε την indegrand(n,x) θα έβλεπε λάθος.

Η quad δεν ξέρει ποια μεταβλητή είναι ποια.
Απλά υποθέτει ότι η συνάρτηση έχει τη μορφή: 
f(x, param1, param2, ...)

```python
from scipy.integrate import quad

def f(x, a, b):
    return a*x**2 + b

result, error = quad(f, 0, 2, args=(2, 1))

print("Integral =", result)
print("Error =", error)
# για args(0,1) 
# για args(1,0) 
```
